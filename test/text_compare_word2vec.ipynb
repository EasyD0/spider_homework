```python
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity

# 定义两个文本
text1 = "This is the first document."
text2 = "This document is the second document."

# 将文本转换为单词列表
tokens1 = text1.lower().split()
tokens2 = text2.lower().split()

# 训练一个 Word2Vec 模型，使用预训练模型也可以
corpus = [tokens1, tokens2]
model = Word2Vec(corpus, window=5, min_count=1, workers=3, sg=0)

# 获取每个单词的向量表示
vectors1 = [model.wv[word] for word in tokens1]
vectors2 = [model.wv[word] for word in tokens2]

# 计算两个文本之间的余弦相似度
similarity = cosine_similarity([vectors1], [vectors2])[0][0]

# 打印相似度
print("相似度：", similarity)
```
